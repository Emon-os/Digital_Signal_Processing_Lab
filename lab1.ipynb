```python
import numpy as np
import matplotlib.pyplot as plt

# Generate n values
n = np.arange(0, 50)   # 0 to 49

# Sine and cosine sequences
sine_seq = np.sin(0.2 * np.pi * n)
cosine_seq = np.cos(0.2 * np.pi * n)

# Plot sine
plt.figure()
plt.stem(n, sine_seq)
plt.title("Sine Sequence")
plt.xlabel("n")
plt.ylabel("sin(0.2πn)")
plt.grid(True)

# Plot cosine
plt.figure()
plt.stem(n, cosine_seq)
plt.title("Cosine Sequence")
plt.xlabel("n")
plt.ylabel("cos(0.2πn)")
plt.grid(True)

plt.show()
```
