```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 51
fc = 0.2
M = N - 1

h = np.zeros(N)

for n in range(N):
    if n == M/2:
        h[n] = 2*fc
    else:
        h[n] = np.sin(2 * np.pi * fc * (n - M/2)) / (np.pi * (n - M/2))

w = np.hamming(N)
h = h * w

H = np.fft.fft(h,1024)
H_mag = np.abs(H[:512])
freq = np.linspace(0,0.5,512)

plt.figure()
plt.plot(freq,H_mag)
plt.title("FIR Low Pass Filter Frequency Response")
plt.xlabel("Normalized Frequency")
plt.ylabel("Magnitude")
plt.grid()
plt.show()
```
