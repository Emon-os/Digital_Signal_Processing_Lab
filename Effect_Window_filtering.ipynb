```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 51
fc = 0.2
M = N-1

# Ideal response
n = np.arange(N)
hd = np.sinc(2*fc*(n-M/2))

# Rectangular window
rect = np.ones(N)

# Hamming window
hamm = np.hamming(N)

h_rect = hd * rect
h_hamm = hd * hamm

# Frequency response
H_rect = np.abs(np.fft.fft(h_rect, 1024))
H_hamm = np.abs(np.fft.fft(h_hamm, 1024))

plt.figure()
plt.plot(H_rect[:512])
plt.title("Rectangular Window Response")
plt.show()

plt.figure()
plt.plot(H_hamm[:512])
plt.title("Hamming Window Response")
plt.show()
```
