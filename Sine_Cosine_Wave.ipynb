```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Sine wave parameters
frequency = 5     #frequency in Hz
amplitude = 1     #Amplitude
phase = 0         #Phase
duration = 1      #Duration in seconds
sampling_rate = 1000 #Samples per second

#Generate time array
t = np.linspace(0,duration, sampling_rate * duration)

#Generate sine wave : y = A * sin(2πft + φ)
y = amplitude * np.sin(2 * np.pi * frequency * t + phase)
```

```python
plt.figure(figsize=(10,4))
plt.plot(t,y,linewidth=1)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title(f'Sine Wave: f = {frequency} Hz, A = {amplitude}')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.xlim(0, duration)
plt.ylim(-amplitude * 1.2, amplitude * 1.2)
plt.tight_layout()
plt.show()
```

```python
# Sine wave parameters
frequency = 5      # Frequency in Hz
amplitude = 1      # Amplitude
phase = 90         # Phase in degrees
duration = 1       # Duration in seconds
sampling_rate = 1000  # Samples per second

# Generate time array
t = np.linspace(0, duration, sampling_rate * duration)

# Generate sine wave: y = A * sin(2πft + φ)
y = amplitude * np.sin(2 * np.pi * frequency * t + phase)

# Plot the sine wave
plt.figure(figsize=(12, 4))
plt.plot(t, y, linewidth=1)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title(f'Sine Wave: f = {frequency} Hz, A = {amplitude}')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.xlim(0, duration)
plt.ylim(-amplitude * 1.2, amplitude * 1.2)
plt.tight_layout()
plt.show()
```

```python
# Cosine wave parameters
frequency = 5          # Frequency in Hz
amplitude = 1          # Amplitude
phase = np.deg2rad(90) # Phase in radians (90°)
duration = 1           # Duration in seconds
sampling_rate = 1000   # Samples per second

# Generate time array
t = np.linspace(0, duration, sampling_rate * duration)

# Generate cosine wave: y = A * cos(2πft + φ)
y = amplitude * np.cos(2 * np.pi * frequency * t + phase)

# Plot the cosine wave
plt.figure(figsize=(12, 4))
plt.plot(t, y, linewidth=1)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title(f'Cosine Wave: f = {frequency} Hz, A = {amplitude}')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, linewidth=0.5)
plt.xlim(0, duration)
plt.ylim(-amplitude * 1.2, amplitude * 1.2)
plt.tight_layout()
plt.show()
```

```python
# Wave parameters
frequency = 5          # Frequency in Hz
amplitude = 1          # Amplitude
phase = np.deg2rad(90) # Phase in radians
duration = 1           # Duration in seconds
sampling_rate = 1000   # Samples per second

# Generate time array
t = np.linspace(0, duration, sampling_rate * duration)

# Generate sine wave
y_sin = amplitude * np.sin(2 * np.pi * frequency * t + phase)

# Generate cosine wave
y_cos = amplitude * np.cos(2 * np.pi * frequency * t + phase)

# Plot both waves
plt.figure(figsize=(12, 4))
plt.plot(t, y_sin, label='Sine Wave', linewidth=1)
plt.plot(t, y_cos, label='Cosine Wave', linewidth=1, linestyle='--')

plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title(f'Sine & Cosine Wave: f = {frequency} Hz, A = {amplitude}')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, linewidth=0.5)
plt.xlim(0, duration)
plt.ylim(-amplitude * 1.2, amplitude * 1.2)
plt.legend()
plt.tight_layout()
plt.show()
```
