```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Sampling
fs = 500
t = np.linspace(0, 1, fs)

# Original signal (Low + High frequency)
x = np.sin(2*np.pi*5*t) + 0.5*np.sin(2*np.pi*50*t)

# Simple FIR Low Pass Filter (Moving Average)
N = 15
h = np.ones(N)/N
y = np.convolve(x, h, mode='same')

# Plot
plt.figure()
plt.plot(t, x)
plt.title("Original Signal (Low + High Frequency)")
plt.show()

plt.figure()
plt.plot(t, y)
plt.title("After Low Pass Filtering")
plt.show()
```
