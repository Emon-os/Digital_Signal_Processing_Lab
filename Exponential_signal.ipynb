```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#time axis
t = np.linspace(-5,5,1000)

#exponential constant
a=1 #change to -1 for decay

x = np.exp(a*t)

#plot
plt.plot(t,x)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Exponential Decay")
plt.grid(True)
plt.show()
```
