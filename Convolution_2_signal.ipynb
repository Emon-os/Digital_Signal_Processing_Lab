```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Input signals
x = np.array([1, 2, 3])
h = np.array([1, 1, 1])

# Convolution
y = np.convolve(x, h)

# Index
n_x = np.arange(len(x))
n_h = np.arange(len(h))
n_y = np.arange(len(y))

# Plot
plt.subplot(3, 1, 1)
plt.stem(n_x, x)
plt.title("Input Signal x[n]")
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(n_h, h)
plt.title("Input Signal h[n]")
plt.grid(True)

plt.subplot(3, 1, 3)
plt.stem(n_y, y)
plt.title("Convolution Output y[n]")
plt.grid(True)

plt.tight_layout()
plt.show()
```
